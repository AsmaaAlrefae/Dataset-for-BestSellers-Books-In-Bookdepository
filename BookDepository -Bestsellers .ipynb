# Loop Through the first ten pages in the Bestsellers section and get the following features :

- Book_title
- Book_author 
- Book_Format 
- Book_publishedDate

```python
Book_title =[]
Book_author =[]
Book_Format =[]
Book_publishedDate =[]
for i in range (1,21):
    print(i)
    response = requests.get("https://www.bookdepository.com/bestsellers?page=%d" %i)
    soup1 = BeautifulSoup(response.text, 'html.parser')
    for i in soup1.find_all("h3",attrs={"class":"title"}):
        Book_title.append(i.text.split("\n\n                    ")[1].split("\n")[0])
    for i in soup1.find_all("p",attrs={"class":"author"}): 
        Book_author.append(i.text.split("\n")[1].split("\n")[0])
    for i in soup1.find_all("p",attrs={"class":"format"}): 
        Book_Format.append(i.text)
    for i in soup1.find_all("p",attrs={"class":"published"}): 
        Book_publishedDate.append(i.text)
    
```

# Create a DataFrame


```python
#Create the dataframe below!
Books_df = pd.DataFrame({"Title":Book_title,"Author":Book_author,"Format":Book_Format,"Published Date" : Book_publishedDate}) 
```
